# Import Modules

```python
import cv2
import matplotlib.pyplot as plt
#%matplotlib inline
```

## HAAR Cascade File Path

```python
# Load a pre-trained Haar Cascade classifier for frontal face detection
# - 'cv2.data.haarcascades' provides access to the folder where OpenCV stores pre-trained Haar Cascade files.
# - 'haarcascade_frontalface_default.xml' is the specific Haar Cascade XML file for detecting faces in an image.
# - This classifier uses Haar features to detect faces based on patterns in pixel intensity that appear in human faces.
#face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + '/kaggle/input/computervision/haarcascade_frontalface_default.xml')

# The 'face_cascade' object now contains the trained classifier, which can be used to detect faces in images or video frames.
```

```python
import os
print(os.path.exists('/kaggle/input/computervision/haarcascade_frontalface_default.xml'))
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

# Load the Image

```python
# Read the image from the specified file path ('test image.jpg')
# 'cv2.imread()' loads the image into memory. The path must be correct for this to work.
image = cv2.imread('/kaggle/input/computervision/test image.jpg')

# Convert the image from BGR (default OpenCV color format) to RGB
# OpenCV loads images in BGR format by default, but matplotlib expects RGB for proper display.
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# 'cv2.cvtColor()' changes the color space from BGR to RGB.

# Display the RGB image using matplotlib
# 'plt.imshow()' renders the image in RGB color format, suitable for display in notebooks.
plt.imshow(img_rgb)
```

```python
# Resize the image to a new dimension (400x600 pixels)
# 'cv2.resize()' is used to change the size of the image to the specified width (400) and height (600).
# This is useful if you need to standardize image sizes for further processing.
image = cv2.resize(image, (400, 600))
```

```python
# Convert the resized image to grayscale
# Grayscale images contain only intensity information (black and white), with no color.
# This is often used in computer vision for simpler analysis, like edge detection or thresholding.
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# 'cv2.cvtColor()' converts the image to grayscale by removing color information.

# Display the grayscale image using matplotlib, with a grayscale colormap
# 'plt.imshow()' is used again, but with the 'cmap='gray'' argument to ensure it displays in black and white.
plt.imshow(gray, cmap='gray')
```

## Detect Faces

```python
# Detect faces in the grayscale image using the previously loaded Haar Cascade classifier
# 'detectMultiScale()' detects objects (faces) in the image. It returns a list of bounding boxes for detected faces.
faces = face_cascade.detectMultiScale(gray)
# This method detects objects (faces) at different scales, which is why it's called 'multi-scale'.
```

```python
# Get the number of detected faces
# 'len(faces)' returns the number of detected faces. If no faces are detected, the length will be 0.
len(faces)
```

```python
# Draw rectangles around the detected faces in the image
# For each face detected, 'faces' contains a bounding box with coordinates (x, y) for the top-left corner and (w, h) for width and height.
for (x, y, w, h) in faces:
    # 'cv2.rectangle()' draws a rectangle around each face detected. The rectangle is drawn in green with thickness 2.
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the image with rectangles drawn around the faces using matplotlib
# 'matplotlib.pyplot.imshow()' is used to display the image instead of cv2.imshow(), which works better in Jupyter environments.
# Convert the image from BGR to RGB before displaying it, as OpenCV uses BGR and matplotlib uses RGB.
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)  # 'plt.imshow()' shows the image with faces outlined by green rectangles.
plt.axis('off')  # Hide axis to make the image look cleaner
plt.show()  # Display the image in the notebook

# Wait for a key press to close the window (for local OpenCV applications)
# Since we are using matplotlib, 'cv2.waitKey()' is not needed in this case.
```

```python
# Import the necessary libraries
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/kaggle/input/computervision/test image.jpg')

# Convert the image to grayscale (required for face detection)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load the Haar Cascade classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Adjust the parameters for better face detection:
# scaleFactor: compensates for the faces appearing smaller/larger at different distances in the image
# minNeighbors: determines the quality of the detection (higher means fewer but more accurate faces detected)
# minSize: defines the smallest face size that should be detected
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Draw rectangles around the detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Convert the image to RGB for display with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with the rectangles drawn around the faces
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# Print the number of faces detected
print(f"Number of faces detected: {len(faces)}")
```
